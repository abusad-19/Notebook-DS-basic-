Hi, I hope you are good. Today, we want to find the answer of the question **Does scaling affet the number of outliers?** <br> Let's start the journey.<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('/content/Housing.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
numericColumnsName=df.select_dtypes(np.number).columns
```

# Before using scaling

```python
#lower & upper bound before droping outliers of "price"
for each in numericColumnsName:
  q1=df[each].quantile(0.25)
  q3=df[each].quantile(0.75)
  iqr=q3-q1
  upperLimit=q3+(1.5*iqr)
  lowerLimit=q1-(1.5*iqr)
  print(f'number of outliers for {each} columns : {df[(df[each]<lowerLimit) | (df[each]>upperLimit)].shape[0]}')
```

# After Using Scaling

```python
mms=MinMaxScaler()
```

```python
df.shape
```

```python
array=mms.fit_transform(df[numericColumnsName])
```

```python
array[0]
```

```python
scaledDf=pd.DataFrame(array,columns=numericColumnsName)
```

```python
#lower & upper bound before droping outliers of "price"
for each in numericColumnsName:
  q1=scaledDf[each].quantile(0.25)
  q3=scaledDf[each].quantile(0.75)
  iqr=q3-q1
  lower=q1-1.5*iqr
  upper=q3+1.5*iqr
  print(f'number of outliers for {each} columns : {scaledDf[(scaledDf[each]<lower) | (scaledDf[each]>upper)].shape[0]}')
```

<br>From above experiment we can say that, **scaling doesn't affect the number of outliers.**<br>
