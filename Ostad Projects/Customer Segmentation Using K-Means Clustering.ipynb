```python
df['total_price']=df['Quantity']*df['UnitPrice']
df
```

```python
bills=df.groupby('CustomerID').sum()[['total_price']]
bills
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df
```

```python
df.info()
```

```python
df[['InvoiceDate']].max()
```

```python
#df['invoice_age']=
df['ages']=(pd.to_datetime('1-1-2012')-df['InvoiceDate']).dt.days
df
```

```python
b=df.groupby('CustomerID').min()[['ages']]
b
```

```python
data=pd.concat([b,freq,bills],axis=1)
data
```

```python
data.columns=['ages','freq','total_price']
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ss=MinMaxScaler()
```

```python
ss.fit_transform(data)
```

```python
x=pd.DataFrame(np.c_[ss.fit_transform(data)])
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
ssc=[]
for k in range(1,10):
    km=KMeans(n_clusters=k)
    km.fit_predict(x)
    ssc.append(km.inertia_)
```

```python
ssc
```

```python
plt.plot(range(1,10),ssc,c='r')
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
labeled=pd.concat([b,freq,bills],axis=1)
labeled['class']=kmeans.labels_
labeled
```

```python
labeled.groupby('class').mean()
```

```python
def mark(y):
    if y['class']==1:
        return 'Diamond'
    elif y['class']==2:
        return 'Gold'
    elif y['class']==0:
        return 'Silver' 
    else: return 'F'
```

```python
labeled['label']=labeled.apply(mark,axis=1)
labeled
```

```python
labeled.groupby(labeled['label']).count()
```

```python
lab2=labeled[labeled['label']=='Diamond']
lab2
```

```python
x1=lab2[['ages','InvoiceNo','total_price']]
```
